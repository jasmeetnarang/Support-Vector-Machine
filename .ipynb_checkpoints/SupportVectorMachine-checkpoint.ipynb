##### Generating linearly and non-linearly separable dataset, implementing Support Vector Machine algorithm and applying it on the generated datasets

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from cvxopt import matrix as cvxopt_matrix
from cvxopt import solvers as cvxopt_solvers
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn import preprocessing
from sklearn.metrics import recall_score, precision_score, f1_score
```

```python
def linearlySeperable():
    X = np.random.rand(1000,2)
    #print(X)
    Y = np.zeros([len(X),1])
    a = -5
    b = 6
    count = 0
    for i in range(len(X)):
        m = a*X[i,0] + b*X[i,1]
        
        if(m < 0):
            Y[count] = -1
        else:
            Y[count] = 1
        
        count = count+1
    #print(Y)
        
    return Y, X
```

```python
def linearlyNonSeperable():
    X = np.random.rand(1000,2)
    Y = np.zeros([len(X),1])
    a = 0.4
    count = 0
    for i in range(len(X)):
        m = (X[i,0]-0.5)**2 + (X[i,1]-0.5)**2 - a**2
        #print(m)
        if(m > 0):
            Y[count] = 1
        else:
            Y[count] = -1
        count = count+1
    return Y, X
```

```python
def splitTestTrain(X,Y):
    m = int(len(X)/3)
    
    test_X = X[:m,:]
    test_Y = Y[:m]
    
    train_X = X[m:,:]
    train_Y = Y[m:]
    
    return train_X, train_Y, test_X, test_Y
    
```

```python
def plot_graph(X,Y):
    for i in range(len(test_Y)):
        if (test_Y[i] == 1):
            l=plt.scatter(test_X[i,0], test_X[i,1], c = 'red' )
        elif(test_Y[i] == -1):
            j=plt.scatter(test_X[i,0], test_X[i,1], c = 'blue')
    l.set_label('class1')
    j.set_label('class2')
    plt.xlim(-0.5,7)
    plt.ylim(-0.5,4)
    plt.legend(loc='upper left')
    plt.show()
```

```python
def polynomialKernel(X,Y,q=2):
    p_matrix = np.zeros([len(X),len(Y)])
    c1 = 0
    for i in X:
        c2=0
        for j in Y:
            x_t = np.transpose(i)
            yx_t = np.dot(x_t,j)
            k = (yx_t+1)**q
            p_matrix[c1,c2] = k
            #p_matrix[c2,c1] = k
            
            c2 = c2+1
        c1 = c1+1
             
    return p_matrix
```

```python
def guassianKernel(X, Y, s):
    sub = X - Y
    norm = np.linalg.norm(sub)
    norm = np.square(norm)
    norm = norm/(2 * s * s)
    k = np.exp(-norm)
    return k
```

```python
def gramMatrix(X1,X2):
    sdev = 0.1
    m1 = len(X1)
    m2 = len(X2)
    c1 = 0
    g_matrix = np.zeros([m1, m2])
    for i in (X1):
        c2 = 0
        for j in (X2):
            val = guassianKernel(i, j, sdev)
            g_matrix[c1,c2] = val
            c2 = c2+1
        c1 = c1+1
    return g_matrix

```

```python
from enum import Enum
class KernelType(Enum):
    none = 0
    guss_kernel = 1
    poly_kernel = 2
```

```python
def CalculateAlpha(X,Y,kernel,soft_margin=False):
    m = len(X)
    n = len(Y)
    XX_T = np.matmul(X,np.transpose(X))
    YY_T = np.matmul(Y,np.transpose(Y))

    if (kernel == KernelType.none):
        P = np.multiply(XX_T,YY_T)
        #P = cvxopt_matrix(P)
    elif(kernel == KernelType.guss_kernel):
        P = np.multiply(gramMatrix(X,X),YY_T)
        #P = cvxopt_matrix(P)
    elif(kernel == KernelType.poly_kernel):
        P = np.multiply(polynomialKernel(X,X,q=2),YY_T)
        
        #print(P)
    P = cvxopt_matrix(P)    
    q = cvxopt_matrix(-np.ones((m,1))) 
    A = Y.reshape(1,n)
    A = cvxopt_matrix(A,(1,n),'d')
    
    b0 = 0.0
    b = cvxopt_matrix(b0)
    #print(np.linalg.matrix_rank(A))    
    #print(np.linalg.matrix_rank(P))
    #print(np.linalg.matrix_rank(q))
    
    if (soft_margin == False):
        G = cvxopt_matrix(-np.eye(m))
        h = cvxopt_matrix(np.zeros((m,1)))
    elif (soft_margin == True):
        c = 10
        G = cvxopt_matrix(np.vstack((np.eye(m)*-1,np.eye(m))))
        h = cvxopt_matrix(np.hstack((np.zeros(m), np.ones(m) * c)))
    
    cvxopt_solvers.options['show_progress'] = False
    cvxopt_solvers.options['abstol'] = 1e-10
    cvxopt_solvers.options['reltol'] = 1e-10
    cvxopt_solvers.options['feastol'] = 1e-10
    
    #print(np.linalg.matrix_rank(G))
    
    sol = cvxopt_solvers.qp(P,q,G,h,A,b,kktsolver='ldl', options={'kktreg':1e-9,'show_progress': False})
    alpha = np.array(sol['x'])
    
    return alpha    
    
```

```python
def getSupVector(X,Y,alpha):
    sv_x = []
    sv_y = []
    
    for i in range(len(X)):
        if (alpha[i] > 1e-200):
            sv_x.append(X[i])
            sv_y.append(Y[i])
            
    return np.asarray(sv_x), np.asarray(sv_y)

```

```python
def getWeight(X,Y,alpha):
    Y = Y.reshape(len(Y),1)
    a = np.multiply(alpha, Y)
    #print(alpha.shape)
    #print(Y.shape)
    #print(a.shape)
    w = np.dot(np.transpose(a),X)
    
    return w
```

```python

def getWeightZero(X,Y,alpha,w):
    S = (alpha > 1e-4).flatten()
    
    w0 = Y[S] - np.dot(X[S],np.transpose(w))
    m = len(Y[S])
    w0 = np.sum(w0)/m
    
    return w0
```

```python
def y_prediction(X1,X2,Y,w,w0,alpha,kernel):
    y_pred = np.zeros([len(X),1])

    if (kernel == KernelType.none):
        y_pred = np.matmul(X2,np.transpose(w))+w0
    elif(kernel == KernelType.guss_kernel):
        Y = Y.reshape(len(Y),1)
        sim = gramMatrix(X1,X2)
        alphaY = np.multiply(Y,alpha)
        #print(sim.shape)
        alphaY_t = np.transpose(alphaY)
        #print(alphaY_t.shape)
        y_pred = np.dot(alphaY_t,sim)
        y_pred = y_pred+w0
        y_pred = np.transpose(y_pred)
    elif(kernel == KernelType.poly_kernel):
        Y = Y.reshape(len(Y),1)
        sim = polynomialKernel(X1,X2,q=2)
        alphaY = np.multiply(Y,alpha)
        alphaY_t = np.transpose(alphaY)
        y_pred = np.dot(alphaY_t,sim)
        y_pred = y_pred+w0
        y_pred = np.transpose(y_pred)
        
    #print(y_pred)
    y_pred[y_pred > 0] = 1
    y_pred[y_pred <= 0] = -1
    
    
    return y_pred

    
```

```python
def plotgraph(X,y_pred,sv_x):
    for i in range(len(y_pred)):
        if (y_pred[i] == 1):
            l=plt.scatter(X[i,0], X[i,1], c = 'red')
        elif(y_pred[i] == -1):
            p=plt.scatter(X[i,0], X[i,1], c = 'blue')
    for j in range(len(sv_x)):
        q=plt.scatter(sv_x[j,0],sv_x[j,1], c = 'green')
    #plt.xlim(-1,10)
    #plt.ylim(-1,10)
    try:
        l.set_label('class1')
    except:
        pass
    try:
        p.set_label('class2')
    except:
        pass
    try:
        q.set_label('sv')
    except:
        pass
    plt.legend(loc="upper right")
    plt.show() 
    
```

```python
def crossValidation(X,Y,kernel,soft_margin):
    k = 2
    
    train_X, train_Y, test_X, test_Y = splitTestTrain(X,Y)
    num_samples = len(train_X)//k
    
    validation_accuracies = []
    
    for fold in range(k):
        val_x = train_X[fold*num_samples:(fold+1)*num_samples]
        val_y = train_Y[fold*num_samples:(fold+1)*num_samples]
        
        train_x1 = train_X[:fold*num_samples,:]
        train_x2 = train_X[(fold+1)*num_samples:,:]
        train_x = np.concatenate((train_x1,train_x2),axis=0)

        train_y1 = train_Y[:fold*num_samples]
        train_y2 = train_Y[(fold+1)*num_samples:]
        train_y = np.concatenate((train_y1,train_y2),axis=0)
        
        
        alpha = CalculateAlpha(train_x,train_y,kernel,soft_margin)
        sv_x,sv_y = getSupVector(train_x,train_y,alpha)
        w = getWeight(train_x,train_y,alpha)
        w0 = getWeightZero(train_x,train_y,alpha,w)
        
        ypred = y_prediction(train_x,val_x,train_y,w,w0,alpha,kernel)
        #print("ypred:",ypred)
        accuracy = accuracy_score(val_y, ypred)
         
        validation_accuracies.append(accuracy)
        
        
    validation_accuracy = np.average(validation_accuracies)
    print("validation accuracy: ", validation_accuracy)
    
    #final test
    alpha = CalculateAlpha(train_X,train_Y,kernel,soft_margin)
    sv_x,sv_y = getSupVector(train_X,train_Y,alpha)
    w_test = getWeight(train_X,train_Y,alpha)
    w0_test = getWeightZero(train_X,train_Y,alpha,w_test)

    y_pred = y_prediction(train_X,test_X,train_Y,w_test,w0_test,alpha,kernel)
    accuracy = accuracy_score(test_Y, y_pred)
    precision = precision_score(y_pred,test_Y,average='weighted')
    recall = recall_score(y_pred,test_Y,average='weighted')
    f1 = f1_score(test_Y, y_pred,average='weighted')
    print("test_accuracy: ",accuracy)
    print("precision score: ",precision)
    print("recall score: ",recall)
    print("f1_score: ",f1)
    cm = confusion_matrix(test_Y, y_pred)
    print(cm)
    
    plotgraph(test_X,y_pred,sv_x)
    #y_prediction(train_X,test_X,train_Y,w,w0,alpha,kernel=KernelType.guss_kernel)
       
```

##### External Dataset

```python
def readFile(filename,s):
    data = pd.read_csv(filename,sep = s, header=None)

    return data
```

```python
def splitColumnXY(df):
    df = df.sample(frac=1).reset_index(drop=True)
    X = df.iloc[:,:-1]
    X = preprocessing.scale(X)
    X = pd.DataFrame(X)
    #last column is Y 
    Y = df.iloc[:,-1]
    
    X = X.to_numpy()
    Y = Y.to_numpy()
    Y[Y == 0] = -1  
    
    return X,Y

```

```python
def cleanData(BC_data):
    BC_data = BC_data.sample(frac=1).reset_index(drop=True)
    X = BC_data.iloc[:,:-1]
    v = X[6].value_counts().idxmax()
    X[6] = X[6].replace('?',str(v))
    X = preprocessing.scale(X)
     #last column is Y 
    Y = BC_data.iloc[:,-1]
    
    Y = Y.to_numpy()
    Y[Y == 2] = -1
    Y[Y == 4] = 1
    
    return X,Y
```

```python
BC_data = readFile('breast-cancer-wisconsin.data',',')
BC_data = BC_data.iloc[:,1:]
X,Y = cleanData(BC_data)    
```
